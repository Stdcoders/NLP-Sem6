# Assignment 5 - Use WordNet to identify semantic relationships such as synonymy, antonymy, and hypernymy in text data

```python
import nltk
from nltk.corpus import wordnet as wn

nltk.download('wordnet')
nltk.download('omw-1.4')
```

# 1. Synonyms for the word - 'delight'

```python
synonyms = set()

for syn in wn.synsets("delight"):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

print("Synonyms:", synonyms)
```

# 2. Antonyms for the word - 'delight'

```python
antonyms = set()

for syn in wn.synsets("delight"):
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Antonyms:", antonyms)
```

# 3. Hypernyms for the word - 'human'

```python
hypernyms = set()

for syn in wn.synsets("human"):
    for hyper in syn.hypernyms():
        hypernyms.add(hyper.name())

print("Hypernyms:", hypernyms)
```

# 4. Hyponyms for the word - 'animal'

```python
hyponyms = set()

for syn in wn.synsets("animal"):
    for hypo in syn.hyponyms():
        hyponyms.add(hypo.name())

print("Hyponyms:", list(hyponyms)[:5])
```

# 5. WordNet on the example sentence

```python
sentence = "The human is very delighted".lower().split()

for word in sentence:
    synsets = wn.synsets(word)
    if synsets:
        print(f"\nWord: {word}")
        print("Synonyms:", [l.name() for l in synsets[0].lemmas()])
        print("Hypernyms:", [h.name() for h in synsets[0].hypernyms()])
```
