```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn
```

```python
def disambiguate_word(sentence, ambiguous_word):
    tokens = word_tokenize(sentence)
    sense = lesk(tokens, ambiguous_word)

    if sense:
        print("Word:", ambiguous_word)
        print("Meaning:", sense.definition())
        print("Example:", sense.examples())
        print("Synset:", sense.name())
    else:
        print("No sense found")
```

```python
sentence1 = "He sat on the bank of the river"
sentence2 = "She deposited money in the bank"

disambiguate_word(sentence1, "bank")
print("\n-----------------\n")
disambiguate_word(sentence2, "bank")
```

```python
def wsd_sentence(sentence):
    tokens = word_tokenize(sentence)

    for word in tokens:
        synsets = wn.synsets(word)

        if len(synsets) > 1:  # ambiguous word
            sense = lesk(tokens, word)

            if sense:
                print(f"\nWord: {word}")
                print("Chosen Meaning:", sense.definition())
```

```python
text = "The bat flew at night and the player hit the ball with a bat"
wsd_sentence(text)
```

```python
def compare_meanings(word):
    synsets = wn.synsets(word)

    print("All possible meanings of:", word, "\n")

    for s in synsets:
        print(s.name(), ":", s.definition())
```

```python
compare_meanings("bat")
```
